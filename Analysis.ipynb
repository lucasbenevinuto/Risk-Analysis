```python
import matplotlib.pyplot as plt
import seaborn as sns

f = plt.figure(figsize=(20, 12))
gs = f.add_gridspec(5, 3)

with sns.axes_style("darkgrid"):
    ax = f.add_subplot(gs[0, 0])
    sns.histplot(data=data, x="Age", bins=50, ax=ax)
    
with sns.axes_style("darkgrid"):
    ax = f.add_subplot(gs[0, 1])
    sns.histplot(data=data, x="AnnualIncome", bins=50, ax=ax)

with sns.axes_style("darkgrid"):
    ax = f.add_subplot(gs[0, 2])
    sns.histplot(data=data, x="CreditScore", bins=50, ax=ax)

with sns.axes_style("darkgrid"):
    ax = f.add_subplot(gs[1, 0])
    sns.histplot(data=data, x="Experience", bins=50, ax=ax)

with sns.axes_style("darkgrid"):
    ax = f.add_subplot(gs[1, 1])
    sns.histplot(data=data, x="LoanAmount", bins=50, ax=ax)

with sns.axes_style("darkgrid"):
    ax = f.add_subplot(gs[1, 2])
    sns.histplot(data=data, x="LoanDuration", bins=50, ax=ax)

with sns.axes_style("darkgrid"):
    ax = f.add_subplot(gs[2, 0])
    sns.histplot(data=data, x="NumberOfDependents", bins=50, ax=ax)

with sns.axes_style("darkgrid"):
    ax = f.add_subplot(gs[2, 1])
    sns.histplot(data=data, x="MonthlyDebtPayments", bins=50, ax=ax)

with sns.axes_style("darkgrid"):
    ax = f.add_subplot(gs[2, 2])
    sns.histplot(data=data, x="CreditCardUtilizationRate", bins=50, ax=ax)

with sns.axes_style("darkgrid"):
    ax = f.add_subplot(gs[3, 0])
    sns.histplot(data=data, x="NumberOfOpenCreditLines", bins=50, ax=ax)

with sns.axes_style("darkgrid"):
    ax = f.add_subplot(gs[3, 1])
    sns.histplot(data=data, x="NumberOfCreditInquiries", bins=50, ax=ax)

with sns.axes_style("darkgrid"):
    ax = f.add_subplot(gs[3, 2])
    sns.histplot(data=data, x="DebtToIncomeRatio", bins=50, ax=ax)

with sns.axes_style("darkgrid"):
    ax = f.add_subplot(gs[4, 0])
    sns.histplot(data=data, x="BankruptcyHistory", bins=50, ax=ax)

with sns.axes_style("darkgrid"):
    ax = f.add_subplot(gs[4, 1])
    sns.histplot(data=data, x="PreviousLoanDefaults", bins=50, ax=ax)

with sns.axes_style("darkgrid"):
    ax = f.add_subplot(gs[4, 2])
    sns.histplot(data=data, x="PaymentHistory", bins=50, ax=ax)

f.suptitle("Histogram of Numerical Features in Loan Approval Dataset", fontsize=18)

plt.tight_layout()
plt.show()
```

```python
data['LoanToIncomeRatio'] = data['LoanAmount'] / data['AnnualIncome']

sns.histplot(data['LoanToIncomeRatio'], kde=True, bins=30)
plt.title('Distribution of Loan-to-Income Ratio')
plt.show()

sns.boxplot(data=data, x='LoanApproved', y='LoanToIncomeRatio')
plt.title('Loan Approval vs Loan-to-Income Ratio')
plt.show()
```
